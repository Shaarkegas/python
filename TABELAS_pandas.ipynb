```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CARREGANDO BASE DE DADOS
base = pd.read_csv('trees.csv')
base.shape
```

```python
#DADOS
base.head()
```

```python
#HISTOGRAMA: h = height
h = np.histogram(base.iloc[:,1],bins = 6)
h
```

```python
#VIZUALIZANDO O HISTOGRAMA COM 6 DIVISOES (bins)
plt.hist(base.iloc[:,1],bins=6)
plt.title('Arvores')
plt.ylabel('Frequencia')
plt.xlabel('Altura')
```

```python
#GRAFICOS DE DENSIDADE ME PYTHON

#HISTOGRAMA COM A LINHA DE DISTRIBUIÃ‡AO DE FREQUENCIA COM 6 DIVISOES
#kde = linha de densidade
sns.histplot(base.iloc[:,1], kde = False, bins=6, color='blue').set(title='Arvores')
```

```python
#DENSIDADE, METODO KDEPLOT PARA DENSIDADE
sns.kdeplot(base.iloc[:,1],color='blue').set(title='Arvores')
```

```python
#DENSIDADE E HISTOGRAMA
sns.histplot(base.iloc[:,1], kde=True, bins=6, color='blue').set(title='Arvore')
```

```python
#GRAFICOS DE DISPERSAO
#considerando volume e a dispersao
plt.scatter(x=base.Girth, y=base.Volume, color='blue',facecolors='none',marker='*')
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```

```python
#GRAFICO DE LINHA CONSIDERANDO O VOLUME E CIRCUNFERENCIA SEM MARKER
plt.plot(base.Girth,base.Volume)
plt.title('Arvore')
plt.ylabel('Volume')
plt.xlabel('Circunferencia')
```

```python
#Grafico de dispersao com "afastamento" dos dados(jitter)
#fit_reg linha de tendencia
sns.regplot(x=base.Girth,y=base.Volume, data=base, x_jitter=0.3, fit_reg= False)
```

```python
#ADICIONANDO LEGENDA EM GRAFICO DE DISPERSAO
co2 = pd.read_csv('co2.csv')
#criando duas variaveis para cada atributo (x = conc e y=uptake)
x = co2.conc
y = co2.uptake
```

```python
#Retorna os valores unicos do atributuo "treatment"
unicos = list(set(co2.Treatment))
unicos
```

```python
#Percorre cada tipo de tratamento e cria um grafico
for i in range(len(unicos)):
    indice = co2.Treatment == unicos[i]
    plt.scatter(x[indice],y[indice], label= unicos[i])
plt.legend(loc = 'lower right')
```

```python
#DIVISAO DA TELA (na mesma imagem produzir varios graficos)
#otimo para comparacoes

#vamos utilizar os dados "trees"
#Criacao da figura, no qual os graficos serao posicionados
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(base.Girth,base.Volume)
plt.subplot(2,2,2)
plt.scatter(base.Girth,base.Height,marker= '.')
plt.subplot(2,2,3)
plt.scatter(base.Height,base.Volume,marker= '*', color='red')
plt.subplot(2,2,4)
plt.hist(base.Volume)
```

```python
#BOXPLOT
#patch_artist = True preenche, showfliers = outliers
plt.boxplot(base.Volume, vert = False, showfliers = False, notch = True, patch_artist = True)
plt.title('Arvores moles')
plt.xlabel('Volume')
```

```python
#dados por linha
plt.boxplot(base)
plt.title('arvores podres')
plt.xlabel('Dados')
```

```python
#GERANDO 3 BOXPLOTS, cada um mostrando informacoes diferentes
plt.boxplot(base.Volume, vert=False)
plt.boxplot(base.Girth,vert=False)
plt.boxplot(base.Height,vert=False)
plt.title('arvores')
plt.xlabel('dados')
```

```python
#GRAFICOS DE BARRA E SETORES(PIZZA)
basei = pd.read_csv('insect.csv')
basei.shape
basei.head()
```

```python
#Agrupamento dos dados baseado no atributo 'spray', contando e somando os registros
agrupadoi = basei.groupby(['spray'])['count'].sum()
agrupadoi
```

```python
#GRAFICO DE BARRAS
agrupadoi.plot.bar(color = 'green')
```

```python
#GRAFICO DE SETORES(PIZZA)
agrupadoi.plot.pie()
```

```python
#ADICIONANDO LEGENDAS A PIZZA
agrupadoi.plot.pie(legend=True)
```
