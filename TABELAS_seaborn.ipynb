```python
baseCO2 = pd.read_csv('co2.csv')
baseCO2.head()
```

```python
#GRAFICOS DE DISPERSAO
sb.scatterplot(data=baseCO2, x= 'conc', y='uptake', hue = 'Type')
```

```python
#Selecao de registros especificos da base de dados (Quebec e Mississipi)
q = baseCO2.loc[baseCO2['Type'] == 'Quebec']
m = baseCO2.loc[baseCO2['Type'] == 'Mississippi']
```

```python
#SUBGRAFICO (1 LINHA E DUAS COLUNAS) MOSTRANDO GRAFICOS SOBRE CADA REGIAO
plt.figure(1)
plt.subplot(1,2,1)
sb.scatterplot(x= q.conc,y = q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
sb.scatterplot(x= m.conc, y= m.uptake).set_title('Mississipi')
plt.tight_layout()
```

```python
#REFRIGERADO OU NAO
ch = baseCO2.loc[baseCO2['Treatment'] == 'chilled']
nc = baseCO2.loc[baseCO2['Treatment'] == 'nonchilled']
```

```python
#GRAFICO MOSTRANDO O TRATAMENTO DO CO2
plt.figure(1)
plt.subplot(1,2,1)
sb.scatterplot(x= ch.conc,y = ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
sb.scatterplot(x= nc.conc, y= nc.uptake).set_title('Non Chilled')
plt.tight_layout()
```
